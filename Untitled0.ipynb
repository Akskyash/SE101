```python
from sklearn import tree
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
path ='/content/survey_num_testdrive.csv'
dataset = pd.read_csv(path)
```

```python
attribut = ['Q1 gender', 'Q2 from', 'Q1 ADDICTED RATE', 'Q4 feel addicted',
       'Q5 red or watery e1 after taking drug?', 'Q6 sleep properly',
       'Q7 weight change', 'Q8 mood swings', ' Q9 financial difficulties',
       'Q10 diseases', 'Q11 disease type', 'Q12 harm yourself',
       'Q13 difficulty  personal relationships ',
       ' Q14 Neglect responsibilities', 'Q15 losing interest',
       ' Q16 isolating  ', 'Q17 FAMILY RELATION', 'Q18 eat foods',
       'Q19 reaction slowed', 'Q20 physical symptoms ', 'Q21 body',
       'Q22 reaction body', 'Q24 doctor ']
```

```python
data = dataset.loc[:, attribut]
target = dataset['Q25 Rate  mental health']
```

```python
dataTraining, dataTesting, targetTraining, targetTesting = train_test_split(data, target, test_size=0.2)
```

```python
modeldata = tree.DecisionTreeClassifier(criterion='entropy')
modeldata = modeldata.fit(dataTraining, targetTraining)
```

```python
plt.subplots(figsize=(20,20))
tree.plot_tree(modeldata, feature names = attribut, class_names = modeldata.classes_, filled = True, fontsize=12)
```
